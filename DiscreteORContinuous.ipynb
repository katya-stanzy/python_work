```python
import pandas as pd
import numpy as np
data = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/cylons/master/data.csv', header=0)
data.tail(4)
```

```python
# loop through all the columns and calculate the UNIQUE values inside
# if there are too many unique values-> CONTINUOUS
# if there are too few unique values-> DISCRETE
```

```python
# Age 18 20 12 18 17 16 92 67 55 45 45 18 24 
# Address ABC. BCD DEF ABC BCD BCD AED APO 
# Money $1million $1.5million $30 $15 $0 NULL NULL
# Pets 0 0 0 1 2 5 0 0 0
# DISCRETE= ? Address, Pets
# CONTINUOUS= ? Age, Money
```

```python
# PRINT all unique values!
# if unique values are less-> discrete feature or label
# if unique values are too much-> continuous feature or label

#if its a label-> DISCRETE -> CLASSIFICATION, CONT.-> Regression!
for col_name in data.columns:
  unique_values = data[col_name].value_counts()#unique_value_counter
  unique_values_counted = unique_values.count()
  print(col_name + ' : ' + str(unique_values_counted) ) 
```
